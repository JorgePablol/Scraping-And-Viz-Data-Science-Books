```python
df = pd.read_csv('DS_BOOKS_USA_CLEAN.csv')
df = df.drop_duplicates(subset='title', keep='first')
```

## Best books

```python
df_best = df.query('stars >= 4.5 and reviewers > 2')
```

```python
df_best
```

### High rated and most reviewed

```python
df_hrpop = df.query('stars >= 4.5 and reviewers > 100')
```

```python
df_hrpop.reset_index()
```

```python
df['stars'].value_counts()
```

## Worst books

```python
df_worst = df.query('stars < 4 and reviewers > 0')
```

```python
df_worst.sort_values(by='stars', ascending=True)
```

## Expensive

```python
for x in ['cost', 'cost_2', 'kindle','kindle_2', 'kindle_3']:
    print('Actual column {}'.format(x))
    print(df[x].value_counts())
    print('*'*50)
```
